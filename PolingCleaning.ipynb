```python
# Create dicts for the arugments
usecols = {
 "43-45": "A,B,E,G,I,K",
 "45-50": "A,B,E,G,I,K,U,AF,AG",
 "50-51": "A,B,E,G,I,K,T,U,AB,AC",
 "51-55": "A,B,E,G,I,K,Q,AB,AC",
 "55-59": "A,B,E,G,I,K,Q,AB,AC",
 "59-64": "A,B,E,G,I,K,Q,AB,AC",
 "64-66": "A,B,E,G,I,K,T,AB,AC",
 "66-70": "A,B,E,G,I,K,T,AB,AC",
 "70-74": "A,B,E,G,I,K,Q,R,AB,AC",
 "74": "A,B,E,G,I,K,T,U,AB,AC",
 "74-79": "A,B,E,G,I,K,T,U,AB,AC",
 "79-83": "A,B,E,G,I,K,Q,R,AB,AC",
 "83-87": "A,B,E,G,I,K,U,V,AF,AG",
 "87-92": "A,B,E,G,I,K,M,O,U,V,X,Y,AF,AG",
 "92-97": "A,B,E,G,I,K,S,T,V,W,X,Y,AC,AD,AE,AG,AH",
 "97-01": "A,B,E,G,I,K,R,S,U,V,X,Y,AB,AC,AD,AF,AG",
 "01-05": "A,B,F,H,J,L,W,X,Z,AA,X,Y,AB,AC,AD,AG,AH,AI,AK,AL",
 "05-10": "A,B,F,H,J,K,M,O,P,R,T,U,AC,AD,AF,AG,AI,AJ,AM,AN,AO,AR,AS",
 "10-15": "A,B,F,H,J,L,W,X,Z,AA,X,AC,AD,AF,AG,AI,AJ,AM,AN,AO,AP,AR,AS",
 "15-17": "A,B,F,H,J,L,N,P,X,Y,AA,AB,AD,AE,AG,AH,AJ,AK,AN,AO,AP,AQ,AS,AT",
 "17-19": "A,B,F,H,J,L,N,P,R,T,AB,AC,AE,AF,AH,AJ,AI,AK,AL,AN,AO,AQ,AR,AT,AU,AX,AY,AZ,BA,BC,BD",
 "19-24": "A,B,F,H,J,L,N,P,R,Z,AA,AC,AD,AF,AG,AI,AJ,AL,AM,AP,AQ,AR,AS,AT,AV,AW", 
}

skiprows = {
 "43-45": [12],
 "45-50": [51],
 "50-51": [35],
 "51-55": [53],
 "55-59": [84],
 "59-64": [160],
 "64-66": [68],
 "66-70": [199],
 "70-74": [182],
 "74": [54],
 "74-79": [200],
 "79-83": [221],
 "83-87": [320],
 "87-92": [570],
 "92-97": [411],
 "97-01": [286],
 "01-05": [314],
 "05-10": [690],
 "10-15": [2097],
 "15-17": [328],
 "17-19": [571],
 "19-24": [1984],
}
    
# List of sheet names to ignore
sheets_to_ignore = {"Introduction", "Notes", "24-", "Monthly average", "By election polls", "2014 Euros", "YouGov rolling ratings 2006-7"} 

df_dict_flex = {}
for sheet_name in polling.sheet_names:    
    if sheet_name in sheets_to_ignore:
        continue

    # Iterate over usecols keys and match them with sheet names
    matching_cols = None
    matching_skiprows = None
    for key in usecols:
        if key in sheet_name:  # Adjust this if needed
            matching_cols = usecols[key]
            matching_skiprows = skiprows.get(key)
            break

    # Parse the sheet only if there's a matching column set
    if matching_cols and matching_skiprows is not None:
        df_dict_flex[sheet_name] = polling.parse(sheet_name, usecols=matching_cols, skiprows=matching_skiprows)

# Merge sheets
df_merged = pd.concat(df_dict_flex.values(), ignore_index=True)

# Save to CSV
df_merged.to_csv('Polling_CLEANED1.csv',index=False)
```
